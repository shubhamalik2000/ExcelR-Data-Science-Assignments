## Question 4

```python
from scipy.stats import binom
```

```python
a = binom.pmf(1, 5, (1/200))
a  #probability of exactly one wrong phone call
```

```python
sum = 0
for i in [1,2,3,4,5]:
    sum = sum + binom.pmf(i, 5, (1/200))
sum  #probability of at least one wrong call
# at least one means, 1 or more than 1
```

## Question 5

```python
# Given values and their probabilities
values = [-2000, -1000, 0, 1000, 2000, 3000]
probabilities = [0.1, 0.1, 0.2, 0.2, 0.3, 0.1]

# Calculate the expected value 
expected_value = 0
for x, p in zip(values, probabilities):
    expected_value = expected_value + (x*p)

print("The expected value is:", expected_value)
```

```python
expected_value = 0
for x, p in zip(values, probabilities):
    expected_value = expected_value + (x*p)

sum = 0
for x, p in zip(values, probabilities):
    sum = sum + (x**2 * p)

variance = sum - expected_value**2
print("The expected value is:", variance)
```
